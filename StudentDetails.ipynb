```python
# First connect XAMPP i.e., RUN cell-1 or cell-2

%pip install tk
import tkinter as tk
from tkinter import *
from tkinter import ttk
from tkinter.ttk import Notebook, Style
from tkinter.messagebox import showerror, showinfo

def view_submit_clicked(i):
    try:
        k = int(i)
        try:
            mycursor.execute(f"SELECT * FROM details WHERE Id = {k}")
            myresult = mycursor.fetchall()
            for x in myresult: 
                z = f'''Name : {x[1]} {x[2]}
Aadhaar : {x[3]}
Father : {x[4]}
Mother : {x[5]}
Phone : {x[6]}
Class : {x[7]}
Section : {x[8]}
Payment : {x[9]}'''
            tit=f'Details of id = {k}'
            showinfo(
            title=tit,
            message=z)
        except UnboundLocalError:
            showerror(title='Error',message=f'''
        Invalid Id
                            ''')
    except ValueError:
        showerror(title='Error',message=f'''
    Id should be number
                            ''')
    
def vButton():
    vW = Toplevel()
    vW.title('View Details')
    vW.geometry("640x480")
    vW.resizable(False, False)
    Id=tk.StringVar()
    Id_label = ttk.Label(vW, text="Enter existing Id:")
    Id_label.place(x=10, y=100)
    Id_entry = ttk.Entry(vW,textvariable=Id)
    Id_entry.place(x=100, y=100)
    Id_entry.focus()
    Submit_button = ttk.Button(vW, text="Submit", command=lambda:view_submit_clicked(Id.get()))
    Submit_button.place(x=100, y=150)

def update_submit_clicked(i,j):
    try:
        k = int(i)
        c=[]
        mycursor.execute('SELECT * FROM marks')
        myresult = mycursor.fetchall()
        for x in myresult: 
            c.append(x[0])
        try:
            l = j.upper()
            if l=='PAID':
                if k in c:
                    try:
                        mycursor.execute(f'''UPDATE details SET Payment = '{l}' WHERE Id = {k}''')
                        z=f'Payment details of Id = {k} is updated'
                        tit=f'Payment Updation'
                        showinfo(
                        title=tit,
                        message=z)
                    except UnboundLocalError:
                        showerror(title='Error',message=f'''
            Invalid Id
                                ''')
                else:
                    showerror(title='Error',message=f'''
    Id is not member of the instuition 
                                ''')
            else:
                showerror(title='Error',message=f'''
    Enter PAID (or) NOT PAID
                                ''')
        except AttributeError:
            showerror(title='Error',message=f'''
    Don't Enter  numbers or Symbols
                                ''')
    except ValueError:
        showerror(title='Error',message=f'''
    Id should be Number 
                                ''')

def uButton():
    uW = Toplevel()
    uW.title('Update Payment')
    uW.geometry("640x480")
    uW.resizable(False, False)
    Id=tk.StringVar()
    Id_label = ttk.Label(uW, text="Enter existing Id:")
    Id_label.place(x=10, y=100)
    Id_entry = ttk.Entry(uW,textvariable=Id)
    Id_entry.place(x=150, y=100)
    Id_entry.focus()
    Paid=tk.StringVar()
    Paid_label = ttk.Label(uW, text="Type PAID or NOT PAID : ")
    Paid_label.place(x=10, y=200)
    Paid_entry = ttk.Entry(uW,textvariable=Paid)
    Paid_entry.place(x=150, y=200)
    Paid_entry.focus()
    Submit_button = ttk.Button(uW, text="Submit", command=lambda:update_submit_clicked(Id.get(),Paid.get()))
    Submit_button.place(x=150, y=250)

root = tk.Tk()
root.title("Welcome User")
root.geometry("640x480")
root.resizable(False, False)
style = Style()
style.theme_use('default')
style.configure('TNotebook.Tab', background="Red")
style.map("TNotebook", background= [("selected", "red")])
note =Notebook(root)
frame1= Frame(note, width= 600, height=450)
note.add(frame1, text= 'Admin')
frame2 = Frame(note, width= 600, height=450)
note.add(frame2, text= "Teacher")
note.pack(expand= True, fill=BOTH)

view_button = ttk.Button(frame1,text='View Details',command=vButton)
view_button.place(x=250, y=150)
update_button = ttk.Button(frame1,text='Update Details',command=uButton)
update_button.place(x=243, y=270)
        
def submit_clicked(i):
    try:
        k = int(i)
        try:
            mycursor.execute(f"SELECT * FROM marks WHERE Id = {k}")
            myr = mycursor.fetchall()
            for x in myr:
                z=f'''      Quarterly 
Math : {x[1]}
Science : {x[4]}
Social : {x[7]}
English : {x[10]}
        Half-Yearly
Math : {x[2]}
Science : {x[5]}
Social : {x[8]}
English : {x[11]}
        Annual
Math : {x[3]}
Science : {x[6]}
Social : {x[9]}
English : {x[12]}'''
            tit=f'Marks of id = {k}'
            showinfo(
            title=tit,
            message=z)
        except UnboundLocalError:
            showerror(title='Error',message=f'''
        Invalid Id
                            ''')
    except ValueError:
        showerror(title='Error',message=f'''
    Id should be number
                            ''')
    
def enterMarks():
    Id=tk.StringVar()
    Id_label = ttk.Label(frame2, text="Enter existing Id:")
    Id_label.place(x=10, y=100)
    Id_entry = ttk.Entry(frame2,textvariable=Id)
    Id_entry.place(x=100, y=100)
    Id_entry.focus()
    Submit_button = ttk.Button(frame2, text="Submit", command=lambda:submit_clicked(Id.get()))
    Submit_button.place(x=100, y=150)  
        
enterMarks()

root.mainloop()
```
