```python
#temporary JSON file with userdata
#Run this cell FIRST (everytime after restarting runtime)

import json
existingData= {
  'email':['sherlock@gmail.com','nothing23@yahoo.in','ashok@hotmail.com'],
  'username':['Sherlock','nothing43','Ashok2'],
  'password':['aErO@123','Tilt$34','Ash123$']
}

f = open('existingUserData.json','w')
json.dump(existingData,f)
f.close()
```

```python
#Register - Give unique email, password, unique username (In their respective format)
#Login - Give either existing email (or) existing username and their respective password

import re
import json

#register
def registerUser(registerMail,passkey):
  print(f'Username must contain atleast 6 characters and a maximum of 15 characters,\n should start with a alphabets, should only contain alphabets and number')
  newUser=input('Username :')
  if newUser in existingData['username']: 
    print(f'Username already taken - Give a new username')
    registerUser(registerMail,passkey)
  elif re.match(r'\b[A-Za-z]', newUser) and newUser.isalnum() and 5<len(newUser)<16:
    print(f'welcome {newUser}')
    a=open('existingUserData.json','a')
    existingData['email'].append(registerMail)
    existingData['password'].append(passkey)
    existingData['username'].append(newUser)
    a.close()
  else:
    print('Invalid Username')
    registerUser(registerMail,passkey)

def registerPassword(registerMail):
  print(f'Password must contain atleast 6 characters and a maximum of 15 characters,5\nAtleast one digit, one UPPERCASE , one lowercase , one special character')
  passkey=input('Password : ')
  passwordPattern=re.compile(patternPassword)
  if re.search(passwordPattern, passkey):
    registerUser(registerMail,passkey)
  else:
    print ('Invalid password')
    registerPassword(registerMail)

def registration():
  registerMail=input('Email : ')              # USE ONLY 1 @ in email
  if registerMail in existingData['email']:   # Atleast 2 letters before @ in email
    print('Account already exists ')
    choice = input('Do you want to login - Select Yes / No : ').upper()
    if choice == 'YES':
      login()
    elif choice == 'NO':
      pass
  elif re.match(patternEmail, registerMail):
    registerPassword(registerMail)
  else:
    print(f"Invalid email")
    registration()

#login
def new(i):
  print(f'Password must contain atleast 6 characters and a maximum of 15 characters, \nAtleast one digit, one UPPERCASE , one lowercase , one special character')
  newPassword=input('New Password : ')
  passwordPattern=re.compile(patternPassword)
  if re.search(passwordPattern, newPassword):
    o= open('existingUserData.json','w') 
    existingData['password'][i]=newPassword
    o.close()
    print('Rerouting to login page')
    login()
  else:
    print ('Invalid password')
    new(i)

def oldNew(i):
    oldNewPassword=input('Do want to see your old password or create new password- select Old / New : ').upper()
    if oldNewPassword == 'OLD':
      print('Your old password is ', existingData['password'][i])
      print('Rerouting to login page')
      login()
    elif oldNewPassword == 'NEW':
      new(i) 
    else:
      print('Invalid Selection')
      oldNew(i)

def mailVerification(a,b):
  i=existingData['email'].index(a)
  if b==existingData['password'][i]:
    welcome=existingData['username'][i]
    print(f'Welcome back, {welcome}')
  else:
    choice=input(f'Incorrect Password \nForget Password - select Yes / No : ').upper()
    if choice == 'YES':
      oldNew(i)
    elif choice == 'NO':
      print('Rerouting to login page')
      login()


def wrongUser():
  choice=input(f'Account does not exists \n Do you want to register - Yes / No : ').upper()
  if choice == 'YES':
    print('Rerouting to Registration')
    registration()
  elif choice == 'NO':
    print('Rerouting to login page')
    login()

def login():
  loginEmailUser=input('Email / Username : ')
  loginPassword=input('Password : ')
  if '@' in loginEmailUser:
    loginMail=loginEmailUser
    if loginMail in existingData['email']:
      mailVerification(loginMail,loginPassword)
    else:
      wrongUser()
  else:
    loginUser=loginEmailUser
    if loginUser in existingData['username']:   
      j=existingData['username'].index(loginUser)
      userMail=existingData['email'][j]
      mailVerification(userMail,loginPassword)
    else:
      wrongUser()

def selection():
    registerLogin = input('Select Login / Register : ').upper()
    if registerLogin == 'LOGIN':
      return login()
    elif registerLogin == 'REGISTER':
      return registration()
    else:
      print('Invalid Selection')
      return selection()

r = open('existingUserData.json')
r.read()
patternPassword=r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[!\#$%&\'"()*+,-./:;<=>?[\]^_`{|}~@])[A-Za-z0-9!\#$%&\'"()*+,-./:;<=>?[\]^_`{|}~@]{6,15}$'
patternEmail=r'\b[A-Za-z]+[A-Za-z0-9!\#$%&\'"()*+,-./:;<=>?[\]^_`{|}~]+@+[A-Za-z]+\.[A-Za-z]{2,}\b'
selection()
r.close()
```
